```python
# Alexander Domagala
# ECE 6560 - Final Project - Linear Heat Equation
# Spring 2024

# Performs image smoothing based on the Linear Heat
# equation (quadratic penalty)

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# load image
start_pix = 300
end_pix = start_pix+100
img = Image.open('../images/boats.bmp')
img = np.array(img)
img = img[start_pix:end_pix,start_pix:end_pix]

# display image
print(img.shape)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
# add noise to image
mean = 0
std_dev = 20
noise = np.random.normal(mean, std_dev, (img.shape[0],img.shape[1]))
noisy_img = img + noise

# display image
print(noisy_img.shape)
plt.imshow(noisy_img, cmap='gray')
plt.show()
```

```python
def linearHeat(I, timestep, iterations, skip):

  # iterate through all the pixels in the image
  for iter in range(iterations):
    for i in range(1,I.shape[0]-1):
      for j in range(1,I.shape[1]-1):
        I[i,j] = I[i,j] + timestep * ((I[i+1,j] - 2*I[i,j] + I[i-1,j]) + (I[i,j+1] - 2*I[i,j] + I[i,j-1]))
    
    if (iter % skip == 0):
      print(iter)
      plt.imshow(I,cmap='gray')
      plt.show()
```

```python
# run test
timestep = .05
iterations = 200
skip = 10

linearHeat(noisy_img, timestep, iterations, skip)
```
