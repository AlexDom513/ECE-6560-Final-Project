```python
# Alex Domagala
# ECE 6560

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open('../images/boats.bmp')
img = np.array(img)
img = img[300:400,300:400]
print(img.shape)
noise = 30*np.random.randn(img.shape[0],img.shape[1]) # standard normal distribution
print(noise.shape)
plt.imshow(img,cmap='gray')
plt.show()
img = img + noise
plt.imshow(img, cmap='gray')
plt.show()
```

```python
def linearHeat(I):
  Inew = I
  timestep = .1

  # iterate through all the pixels in the image
  for k in range(150):
    for i in range(1,I.shape[0]-1):
      for j in range(1,I.shape[1]-1):
        Inew[i,j] = Inew[i,j] + timestep*((Inew[i+1,j] - 2*Inew[i,j] + Inew[i-1,j]) + (Inew[i,j+1] - 2*Inew[i,j] + Inew[i,j-1]))
    if (k % 1 == 0):
      print(k)
      plt.imshow(Inew,cmap='gray')
      plt.show()

print(img.shape)
linearHeat(img)
```
